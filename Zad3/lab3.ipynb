```python
#zad 1 
def lot(paliwo : int, spalanie: int, czas: int):
    print(f'Paliwo: {paliwo}, Spalanie: {spalanie}, Czas: {czas}')
    while(paliwo>0):
        paliwo = paliwo - spalanie
        czas = czas + 1 
        print(f'Paliwo: {paliwo}, Spalanie: {spalanie}, Czas: {czas}')

lot(100,10,0)
```

```python
#zad 2
def pierwsze():
    liczby = []
    hold = 2
    while len(liczby) < 10:
        for x in range(2,hold):
            if hold % x == 0:
                break
        liczby.append(hold)
        hold=hold + 1
    print(liczby)
        

pierwsze()
```
