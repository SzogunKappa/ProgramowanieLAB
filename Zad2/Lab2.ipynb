```python
#zad 1
def egzamin(punkty: int):
    if punkty > 80:
        print("Zdałeś")
        return 1
    if punkty > 50:
        print("Możesz poprawić")
        return 0 
    print("Musisz poprawić")
    return -1

egzamin(90)
egzamin(66)
egzamin(10)
```

```python
#zad 2
def sort(x: int,y: int,z: int):
    if x>y:
        x, y = y, x
    if x > z:
        x, z = z, x
    if y > z:
        y, z = z, y
    return x, y, z
print(sort(1,2,3))
print(sort(1,3,2))
print(sort(2,1,3))
print(sort(2,3,1))
print(sort(3,1,2))
print(sort(3,2,1))
```

```python
#zad 5
def dopisanie(file, text):
    with open(file,"a") as plik:
        plik.write("\n"+text)

    with open(file,'r') as plik:
        print(plik.read())
dopisanie("notowania_gieldowe.txt","ALR, 113")
        
```

```python
#zad 6
def letter_check(letter):
    if len(letter) == 1 and letter.isalpha():
        if letter.isupper():
            print(f"{letter} jest wielką literą")
            return 0
        print(f"{letter} jest małą literą")
        return 1
    print("tylko jedna litera jest dozwolona")
    return -1

letter_check("A")
letter_check("c")
letter_check("5")
letter_check("\"")
letter_check("AB")
```

```python
#zad 7 
def password_check(password):
    if len(password) == 11 and "!" in password:
        print(password)
        return 0
    print("Hasło jest nie poprawne")
    return -1

password_check("abcdefghij!")
password_check("abcdefghijk")
Hasło = 'pk47!jy0893'
password_check(Hasło)
```
